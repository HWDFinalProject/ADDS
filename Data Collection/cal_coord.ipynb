# 그리드 좌표 계산

```python
# 순서대로 북동, 북서, 남동, 남서 꼭짓점의 위도/경도
NE_lat, NE_lon = 37.7222452, 127.2165021
NW_lat, NW_lon = 37.7222452, 126.7312130
SE_lat, SE_lon = 37.4024968, 126.7312130
SW_lat, SW_lon = 37.4024968, 127.2165021
```

```python
# 이동값 계산
range_lat = NE_lat - SE_lat
range_lon = NE_lon - NW_lon

print(f'위도 이동: {range_lat}°')
print(f"경도 이동: {range_lon}°")
```

```python
# 실제 그리드 크기: 공간 해상도 * 해상도
grid_width = 0.51 * 1024
grid_height = 0.51 * 1024

# 좌표 그리드 크기(경도는 1도당 약 88km, 위도는 1도당 약 111km)
grid_lon = grid_width / 88000
grid_lat = grid_height / 111000

print(f"(좌표) 그리드 크기: {grid_lon, grid_lat}")
```

```python
grid_len_width = round(range_lon / grid_lon)
grid_len_height = round(range_lat / grid_lat)

print(f"그리드 개수) 가로 : {grid_len_width}개, 세로 : {grid_len_height}개")
```

```python
# 첫 그리드의 중심점 계산
first_lat, first_lon = NW_lat - (grid_lat / 2), NW_lon + (grid_lon / 2)
print(f"first : {first_lat, first_lon}")
```

```python
lats = [first_lat - i * grid_lat for i in range(grid_len_height)]
lons = [first_lon + i * grid_lon for i in range(grid_len_width)]
```

```python
coords = [(lat, lon) for lat in lats for lon in lons]
```

# 확인

```python
import matplotlib.pyplot as plt

x = [coord[1] for coord in coords]
y = [coord[0] for coord in coords]

plt.figure(figsize = (10, 8))
plt.scatter(x, y, color = 'blue', s = 10)
plt.title('Grids(82 * 68)')
plt.grid(alpha = 0.5)
plt.show()
```
