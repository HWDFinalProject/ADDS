# 매핑
## 1. 데이터 형태
대장: 시도 시군구 동 따로


---
카카오: 다 하나에 (address)

## 2.시도 표시
대장: ['인천광역시' nan '경기도' '서울특별시']


---


카카오: ['경기' nan '서울' '인천']

## 3. 번지
대장:번 = 본번, 지 = 부번  (본번이 0인 경우 결측, 부번이 0인 경우 부번이 없음)


---


카카오:경기 부천시 오정구 대장동 546-9	본번-부번

## 4. 대장 시군구
대장: 시군구에 분류가 2개인 경우 하나에 있음 예시: 부천시 원미구

```python
# ====================
# 📦 필수 라이브러리 임포트
# ====================
import pandas as pd
import numpy as np

# ==========================
# 🛠 데이터 전처리 함수 정의
# ==========================
def split_bunji(x):
    """ 안정화된 번지 분할 함수 """
    if pd.isna(x):
        return (pd.NA, pd.NA)
    if '-' in x:
        parts = x.split('-')
        return (int(parts[0]), int(parts[1]))
    elif x.startswith('산'):
        return (int(x[1:]), pd.NA)
    else:
        try:
            return (int(x), pd.NA)
        except:
            return (pd.NA, pd.NA)

def preprocess_bd(bd_df):
    """ BD 데이터 전처리 (오류 수정 버전) """
    df = bd_df.copy()

    # 시도 매핑
    sido_map = {'경기':'경기도', '서울':'서울특별시', '인천':'인천광역시'}
    df['시도'] = df['address'].str.split().str[0].map(sido_map)

    # 주소 분할
    split_addr = df['address'].str.split(n=3, expand=True)
    df['시군구'] = split_addr[1].str.replace('시$', '', regex=True)
    df['법정동'] = split_addr[2].str.replace('동$', '', regex=True)

    # 번지 추출
    bunji = df['address'].str.extract(r'(\d+-\d+|\d+|산\d+)')[0]
    df[['번','지']] = pd.DataFrame(
        bunji.apply(split_bunji).tolist(),
        index=df.index
    )

    # 타입 변환
    return df.astype({'번':'Int64', '지':'Int64'})

def preprocess_dj(dj_df):
    """ DJ 데이터 전처리 (오류 수정 버전) """
    df = dj_df.copy()

    # 법정동 정규화
    df['법정동'] = df['법정동'].str.strip().str.replace('동$', '', regex=True)

    # 번/지 처리
    df['번'] = df['번'].replace(0, pd.NA).astype('Int64')
    df['지'] = df['지'].replace(0, pd.NA).astype('Int64')

    return df


# ====================
# 🧹 데이터 전처리 실행
# ====================
bd_pre = preprocess_bd(bd)
dj_pre = preprocess_dj(dj)

# ====================
# 🤝 데이터 병합 실행
# ====================
merged = pd.merge(
    bd_pre,
    dj_pre,
    on=['시도', '시군구', '법정동', '번', '지'],
    how='inner',
    suffixes=('_bd', '_dj')
)

# ====================
# 🔍 결과 검증
# ====================
def validate_results(df):
    print("="*40)
    print("✅ 데이터 구조 검증")
    print(f"- 컬럼 수: {len(df.columns)}")
    print(f"- 행 수: {len(df)}")

    print("\n✅ NULL 값 검사")
    print(df.isnull().sum())

    print("\n✅ 데이터 타입 검증")
    print(df.dtypes)

    print("\n✅ 샘플 데이터")
    return df.head()

validate_results(merged)
```

```python
# ====================
# 📦 필수 라이브러리 임포트
# ====================
import pandas as pd
import numpy as np

# ==========================
# 🔄 DJ → BD 형식 변환 함수
# ==========================
def transform_dj_to_bd(dj_df):
    """ DJ 데이터를 BD 주소 형식으로 표준화 """
    df = dj_df.copy()

    # 1. 시도명 단축
    sido_map = {
        '경기도': '경기',
        '서울특별시': '서울',
        '인천광역시': '인천'
    }
    df['시도'] = df['시도'].map(sido_map)

    # 2. 법정동 '동' 추가
    df['법정동'] = df['법정동'].apply(
        lambda x: f"{x}동" if not str(x).endswith('동') else x
    )

    # 3. 번지 생성 (산/블록 구분)
    df['번지'] = df.apply(lambda row:
        f"산{row['번']}" if row['대지구분'] == '산' else
        f"{row['번']}-{row['지']}" if pd.notna(row['지']) and row['지'] != 0 else
        f"{row['번']}",
        axis=1
    )

    # 4. 주소 문자열 생성 (BD 형식)
    df['address'] = df['시도'] + ' ' + df['시군구'] + ' ' + df['법정동'] + ' ' + df['번지']

    return df

# ====================
# 🧩 데이터 병합 로직
# ====================
def merge_datasets(bd, dj):
    # 1. 데이터 로드

    # 2. DJ 형식 변환
    dj_transformed = transform_dj_to_bd(dj)

    # 3. 주소 기준 병합
    merged = pd.merge(
        bd,
        dj_transformed,
        on='address',
        how='inner',
        validate='1:m'
    )

    # 4. 성공률 계산
    success_rate = len(merged)/len(bd)
    print(f"▶ 병합 성공률: {success_rate:.1%}")

    return merged

# ====================
# 🔍 검증 리포트 추가
# ====================
def generate_merge_report(merged_df):
    report = {
        'total_records': len(merged_df),
        'missing_pk': merged_df['PK'].isna().sum(),
        'column_details': dict(merged_df.dtypes),
        'sample_data': merged_df.head(3).to_dict('records')
    }
    return report

# ====================
# 🚀 전체 실행 파이프라인
# ====================
if __name__ == "__main__":
    # 데이터 경로 (실제 경로로 변경 필요)
    merged_data = merge_datasets(
      bd,dj
    )

    # 결과 저장
    merged_data.to_csv("merged_final.csv", index=False)
    print("\n✅ 병합 결과 저장 완료")

    # 리포트 생성
    report = generate_merge_report(merged_data)
    print("\n📊 병합 리포트:")
    for key, val in report.items():
        if key != 'sample_data':
            print(f"- {key}: {val}")
```

주요시설과의 거리 고려

```python
import geopandas as gpd
from shapely.geometry import Point

# CSV A 생성 (과천청사, 팔당댐, 청평댐)
data_a = {
    'name': ['과천정부청사', '팔당댐', '청평댐'],
    'lat': [37.4269905, 37.5254375, 37.7213125],
    'lon': [126.9840938, 127.2784375, 127.4160625]
}

a_gdf = gpd.GeoDataFrame(
    data_a,
    geometry=gpd.points_from_xy(data_a['lon'], data_a['lat']),
    crs='EPSG:4326'
).to_crs('EPSG:32652')  # 한국 UTM 좌표계로 변환

# CSV B 처리
b_gdf = gpd.read_file("/content/merged_final.csv")
b_gdf = gpd.GeoDataFrame(
    b_gdf,
    geometry=gpd.points_from_xy(b_gdf['lon'], b_gdf['lat']),
    crs='EPSG:4326'
).to_crs('EPSG:32652')

# 최종 점수를 저장할 컬럼 초기화
b_gdf['distance_point'] = 0.0

# 각 시설물별로 반복 계산
for _, facility in a_gdf.iterrows():
    # B 데이터와의 거리 계산 (미터 단위)
    distances = b_gdf.geometry.distance(facility.geometry)

    # 3km 이내인 경우 점수 계산
    within_3km = distances <= 3000  # 3km = 3000m
    scores = 3 - (distances[within_3km]/1000)  # km 단위 변환 후 계산

    # 점수 누적
    b_gdf.loc[within_3km, 'distance_point'] += scores

# 결과 저장 (geometry 제외)
b_gdf.drop(columns='geometry').to_csv("final_with_d_score.csv", index=False)
```

# 그리드 분할

```python
import numpy as np

# 원본 격자 생성 (68 x 82)
original_grid = np.arange(1, 68 * 82 + 1).reshape(68, 82)

# 그리드 크기 정의
grid_size = 12

# 결과를 저장할 리스트
grids = []
grid_indices = {}
# 상단 그리드
for j in range(0, 82, grid_size):
    grid = original_grid[:grid_size, j:j+grid_size].flatten().tolist()
    grids.append(grid)

# 하단 그리드
for j in range(0, 82, grid_size):
    grid = original_grid[-grid_size:, j:j+grid_size].flatten().tolist()
    grids.append(grid)

# 좌측 그리드 (상단과 하단 제외)
for i in range(grid_size, 68 - grid_size, grid_size):
    grid = original_grid[i:i+grid_size, :grid_size].flatten().tolist()
    grids.append(grid)

# 우측 그리드 (상단과 하단 제외)
for i in range(grid_size, 68 - grid_size, grid_size):
    grid = original_grid[i:i+grid_size, -grid_size:].flatten().tolist()
    grids.append(grid)

# 결과 확인
for i, grid in enumerate(grids):
    print(f"그리드{i+1}: {grid}")

print(f"총 그리드 수: {len(grids)}")
```

```python
import pandas as pd

# CSV 파일 로드
csv_file = "/content/final_with_d_score.csv"  # 파일 경로 지정
df = pd.read_csv(csv_file)

# 그리드 인덱스 매칭을 위한 딕셔너리 생성
grid_indices = {}

# 모든 그리드를 순회하며 grid_indices에 매핑
for idx, grid in enumerate(grids, start=1):
    for num in grid:
        grid_indices[num] = idx

# `image_name`이 숫자인 경우 정수형 변환 (필요 시)
df['image_name'] = df['image_name'].astype(int)

# `image_name`과 `grid_index` 매칭
df['grid_index'] = df['image_name'].map(grid_indices)

# 그리드별 데이터 분할 및 저장
for idx in df['grid_index'].dropna().unique():
    df_filtered = df[df['grid_index'] == idx]
    df_filtered.to_csv(f"merged_grid_{int(idx)}.csv", index=False)

# 전체 CSV 저장 (grid_index 포함)
df.to_csv("merged_with_grid_index.csv", index=False)

print("그리드 인덱스를 포함한 CSV 파일 생성 완료.")
```

```python
# prompt: merged_grid로 시작하는 csv파일 모두 한데 모아서 zip

import zipfile

def zip_files(file_pattern, zip_filename):
  """Zips all files matching a given pattern into a single zip file."""
  with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
    for file in glob.glob(file_pattern):
      zipf.write(file)
  print(f"Files matching '{file_pattern}' zipped into '{zip_filename}'")

# Example usage: Zip all CSV files starting with "merged_grid"
zip_files("merged_grid_*.csv", "merged_grid_files.zip")
```
